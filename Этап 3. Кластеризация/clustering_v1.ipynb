```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans

from wordcloud import WordCloud
```

```python
def draw_wordcloud(texts, max_words=1000, width=1000, height=500):
    """
    Создание облаков слов
    """
    wordcloud = WordCloud(background_color='white', max_words=50,
                          width=width, height=height)

    #joint_texts = ' '.join(list(texts))
    joint_texts = ' '.join([str(text) for text in texts[:7000]])
    wordcloud.generate(joint_texts)
    return wordcloud.to_image()
```

```python
df = pd.read_csv('/content/drive/MyDrive/text.csv') # Загрузка текста
lst = np.load('/content/drive/MyDrive/embed.npy') # Загрузка эмбеддингов
```

```python
n_clust = 35 # Число кластеров
```

```python
km2 = KMeans(n_clusters=n_clust, random_state=14) # Кластеризация
labels = km2.fit_predict(lst)
```

```python
df['labels'] = labels # Сопоставление текстов и кластеров
```

```python
for i in [27, 29]:
  print(f'Label: {i}')
  display(draw_wordcloud(df['text'][labels == i]))
  print()
```
